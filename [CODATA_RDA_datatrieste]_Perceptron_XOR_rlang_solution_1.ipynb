#Install keras  

```r
install.packages("keras")
```

# Load the needed libraries

```r
library(keras)
```

# Prepare dataset

```r
l1 = c(0,0,0)
l2 = c(0,1,1)
l3 = c(1,0,1)
l4 = c(1,1,0)

xor_truth_table = rbind(l1,l2,l3,l4)
```

```r
xor_truth_table
```

```r
train_x = xor_truth_table[,c(1,2)]
train_y = xor_truth_table[,3]
```

```r
print(train_x)
print(train_y)
```

# Create a sequential model

```r
# Instantiate the model
model <- keras_model_sequential()
```

```r
model <- model %>% layer_dense(units=4, input_shape=c(2), activation="sigmoid") %>% 
  layer_dense(units=1, activation="sigmoid")
```

```r
model %>% compile(loss = "mean_squared_error", optimizer = "sgd")
```

```r
history <- model %>% fit(x = train_x, y = train_y, epochs = 1000)
```

```r
plot(history)
```

```r
print(round(predict(model, train_x)))
```
